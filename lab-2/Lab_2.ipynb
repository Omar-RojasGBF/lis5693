```python
import spacy
```

I installed spaCy and the small English model to get started on some basic text pre-processing.

```python
import requests

url = "https://raw.githubusercontent.com/Omar-RojasGBF/lis5693/refs/heads/main/lab-1/transcript.txt"
response = requests.get(url)
response.raise_for_status()
text = response.text
```

```python
print("Number of characters:", len(text))

print(text[100])
```

I loaded and read my transcript.txt file from my lab-1 GitHub repo. There are 8566 characters, and I loaded the first 100 characters.

```python
nlp = spacy.blank("en")
nlp.add_pipe("sentencizer")

doc = nlp(text)

sentences = list(doc.sents)

print("Number of sentences:", len(sentences))

print("\nFirst 3 sentences:")
for sent in sentences[:3]:
  print(sent)
```

I added a sentencizer to a blank spaCy mdoel. There is 1 sentence in my transcript. The transcript seemed to ignore all periods and other forms of grammar.

```python
words = [token.text.lower() for token in doc if token.is_alpha]

print("Total words:", len(words))
print("Unique words:", len(set(words)))
```

There are 1543 words in total, and 430 unique words.

```python
from collections import Counter

word_freq = Counter(words)

print("Top 10 most frequent words:")
for word, count in word_freq.most_common(10):
  print(word, count)
```

The most frequent non-stopword is "color." This video is about color theory, so I would assume that the word "color" would come up alot, since the speaker could refer to terms such as the color wheel, color swatches, or color palettes.

```python
nlp2 = spacy.load("en_core_web_sm")

doc2 = nlp2(text)

print("Named Entities:")
for ent in doc2.ents:
  print(ent.text, ent.label_)
```

I ran a full spaCy pipeline. There were 9 named entities found. The types of entities that appeared were PERSON, ORG, CARDINAL, ORG, NORP, WORK_OF_ART, DATE, ORDINAL, and GPE.

```python
from spacy.matcher import PhraseMatcher

matcher = PhraseMatcher(nlp2.vocab, attr="LOWER")

phrases = ["color scheme", "color theory", "hues"]

patterns = [nlp2(p) for p in phrases]

matcher.add("TECH_TERMS", patterns)

matches = matcher(doc2)

print("Matches found:")
for match_id, start, end in matches:
  print(doc2[start:end])
  print("Sentence:", doc2[start].sent)
```
