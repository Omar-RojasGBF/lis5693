My search query for this dataset was "cancer." I want to see if I would articles related to cancer research, as well as what specific words/text might pop up among them. I wanted to see if there are certain publishers that make up for most of the articles surrounding cancer research.

```python

import pandas as pd

df = pd.read_csv(io.StringIO(text))
print(df.head(10))
```

```python
df = df[["Publication Year", "Publication Type", "Source Title"]]
df.columns = df.columns.str.replace(' ', '_')
df
```

The columns "Publication Year", "Publication Type", and "Source Title" were chosen because they could help readers know which articles over cancer are newest or from a particular source.

```python

docs = df.Source_Title.tolist()
print(docs[0])
```

```python
stop_words = stopwords.words('english')
```

```python
def clean_doc(doc):
    no_punct = ''
    for c in doc:
        if c not in string.punctuation:
            no_punct = no_punct+c

    words = no_punct.lower().split()

    final_words = []
    for word in words:
        if word not in stop_words:
            final_words.append(word)

    return final_words
```

```python
custom_stopwords = ['noted', 'discharge']
stop_words.extend(custom_stopwords)
print(f"New stopwords added: {custom_stopwords}")
```

```python
cleaned_docs = [clean_doc(doc) for doc in docs]
id2word = corpora.Dictionary(cleaned_docs)
corpus = [id2word.doc2bow(cleaned_doc) for cleaned_doc in cleaned_docs]

print("Cleaning process, ID-Word Index, and Corpus have been regenerated with new stopwords.")

cleaned_with_custom_stopwords = clean_doc(docs[0])
print("\nOriginal document (first one):")
print(docs[0])
print("\nCleaned document with custom stopwords:")
print(cleaned_with_custom_stopwords)
```

```python
lda_model = LdaModel(corpus=corpus, id2word=id2word, num_topics=20)
```

```python
topic_data = []
target_topic_ids = [2, 5, 10, 15]

for topic_id in target_topic_ids:
    # Retrieve the top 10 words and their probabilities for the current topic
    topic_terms = lda_model.show_topic(topic_id, topn=10)

    for word, prob in topic_terms:
        topic_data.append({
            'Topic ID': topic_id,
            'Word': word,
            'Probability': prob
        })

df_topic_words = pd.DataFrame(topic_data)
print("DataFrame 'df_topic_words' created successfully with topic words and probabilities.")
print(df_topic_words.head())
```
