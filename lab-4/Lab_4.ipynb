```python
import requests
import io

url = "https://raw.githubusercontent.com/Omar-RojasGBF/lis5693/refs/heads/main/lab-4/IMDB_Movies.csv"
response = requests.get(url)
response.raise_for_status()
text = response.text
```

I loaded and read my IMDB Movie Reviews Dataset. I saved it as a dataframe as well.

```python
import pandas as pd

df = pd.read_csv(io.StringIO(text))
print(df.head())
```

```python
features = df['review']
target = df['sentiment']
```

"review" and "sentiment" were the only two columns that appeared, so I can examine which reviews are interpreted as being too sentimental or not.

```python
from sklearn.model_selection import train_test_split
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.feature_extraction.text import TfidfTransformer
from sklearn.pipeline import Pipeline

X_train, X_test, y_train, y_test = train_test_split(features, target, test_size=0.3, random_state=42)

print("Training set has {} samples.".format(X_train.shape[0]))
print("Testing set has {} samples.".format(X_test.shape[0]))

X_train = X_train.fillna('')
X_test = X_test.fillna('')

preprocessing = Pipeline([
    ('vect', CountVectorizer()),
    ('tfidf', TfidfTransformer())
])

print("Preprocessing training data...")
train_preprocessed = preprocessing.fit_transform(X_train)

print("Preprocessing test data...")
test_preprocessed = preprocessing.transform(X_test)
```

```python
y_test.unique()
```

I split the dataset into a 70:30 ratio and performed text preprocessing.

```python
from sklearn.naive_bayes import MultinomialNB
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier

nb_classifier = MultinomialNB()
lr_classifier = LogisticRegression(multi_class="ovr")
random_classifier = RandomForestClassifier()

print("Training Naive Bayes classifier...")
nb_classifier.fit(train_preprocessed, y_train)

print("Training Logistic Regression classifier...")
lr_classifier.fit(train_preprocessed, y_train)

print("Training Random Forest classifier...")
random_classifier.fit(train_preprocessed, y_train)
```

```python
nb_predictions = nb_classifier.predict(test_preprocessed)
lr_predictions = lr_classifier.predict(test_preprocessed)
random_predictions = random_classifier.predict(test_preprocessed)
```

```python
import numpy as np

print("NB Accuracy:", np.mean(nb_predictions == y_test))
print("LR Accuracy:", np.mean(lr_predictions == y_test))
print("RF Accuracy:", np.mean(random_predictions == y_test))
```

The findings indicate that the Logistic Regression Classifier was the best performance, with an accuracy of 89%.
